```python
#Fixing missing data

missing_data = dirty_df.columns[dirty_df.isna().any()]
print('columns with missing datas are: ', end=' ')
[print(missing, end=', ') for missing in missing_data]
print('\nSum of missing values in each column: \n')

for column in missing_data:
    print(f'{column} = {dirty_df[column].isnull().sum()}')

for column in missing_data:
    if dirty_df[column].dtype!='object':
        value = dirty_df[column].mean()
    else:
        value = dirty_df[column].mode()
        value = value[0]
    df[column].fillna(value,inplace=True)    
```

```python
# Inconsistencies 

import random

#columns = dirty_df.select_dtypes(include=['object'])
#for column in columns:
#    print(f"{column} {dirty_df[column].unique()}")

#print('\n')


# Since there is no inconsistencies, we will create them
num = 1000
change = dirty_df[dirty_df['Internet_Access']=='Yes'].index
inconsistant = random.sample(list(change), num)
dirty_df.loc[inconsistant, 'Internet_Access'] = 'yes'

print('Internet_Access before fixing inconsistencies')
print(dirty_df['Internet_Access'].head())
print('\n')

# Fixing inconsistencies
dirty_df["Internet_Access"] = dirty_df["Internet_Access"].apply(lambda x: x.replace("yes","Yes"))

print('Internet_Access after fixing inconsistencies')
print(dirty_df['Internet_Access'].head())
```

```python
# Outliers

print('Columns with outlier values')
for column in dirty_df.select_dtypes(include=['int64', 'float64']):
    q1 = dirty_df[column].quantile(0.25)
    q3 = dirty_df[column].quantile(0.75)
    iqr = q3 - q1

    upper = q3 + 1.5*iqr
    lower = q1 - 1.5*iqr
    
    mask = (dirty_df[column] > upper) | (dirty_df[column] < lower)
    print(column)
    df = df[~mask]
    
plt.figure(figsize=(10, 5))    
sns.boxplot(data = dirty_df)
plt.title('Data before removing outliers')
plt.show()

plt.figure(figsize=(10, 5))    
sns.boxplot(data = df)
plt.title('Data after removing outliers')
plt.show()
```

### a. Estimate the unknown model parameters (fitting) and evaluate the model(validation/cross-validation)

```python
#Read and describe the data.
import numpy as np
import pandas as pd

df = pd.read_csv('StudentPerformanceFactors.csv', delimiter =',')
display(df.describe().T)
```

```python
#Calculate the correlation among all the variables.
corr = df.corr()

corr.style.background_gradient(cmap='coolwarm').set_precision(4)

display(corr)

import seaborn as sns
%matplotlib inline

plt.figure(figsize=(8, 6))
sns.heatmap(corr)
plt.title("correlation among all the variables")
plt.show()
```

```python
#Identify the top three correlated input variables to the output variable.

corr=corr.apply(lambda x: np.abs(x))

sorted_corr = corr.sort_values(by=['Exam_Score'], ascending=False) # negative values will be considered using abs.

sorted_corr = sorted_corr['Exam_Score'].index
print('The top three correlated input variables are: ',sorted_corr[1:4].tolist())
print(sorted_corr)
```

```python
#Calculate the coefficient estimates using OLS closed form.
df_encoded = pd.get_dummies(df, drop_first=True)

Xo = df_encoded.iloc[:,0:-1].values
y = df_encoded.iloc[:,-1].values
X = np.c_[np.ones(len(y)), Xo]

best_beta = np.linalg.inv(X.T @ X) @ X.T @ y

print('The closed form estimates are:', np.round(best_beta,2).tolist()) 
```

```python
#Calculate the coefficient estimates using scikit learn LinearRegression module. 
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(Xo, y)

best_beta =  np.round(reg.coef_,2)
best_beta_0 = np.round(reg.intercept_,2)

print(f'The best values for the estimates are :', best_beta_0, best_beta.tolist())
```

```python
# Load the data
import pandas as pd
# basic packages for model calculations
import pandas as pd
import numpy as np

#model importing
from sklearn.linear_model import LinearRegression

# to split the data set into train and test sets and to do cross validation
from sklearn.model_selection import train_test_split, cross_val_score

#These are for evaluating the model accuracy
from sklearn.metrics import mean_squared_error, r2_score

# Import preprocessing for one-hot encoding
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('StudentPerformanceFactors.csv')

# Identify categorical features
categorical_features = df.select_dtypes(include=['object']).columns

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for compatibility with LinearRegression

# Fit the encoder to the categorical features and transform them
encoded_features = encoder.fit_transform(df[categorical_features])

# Create a DataFrame from the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))

# Concatenate the encoded features with the numerical features
numerical_features = df.drop(categorical_features.union(['Exam_Score']), axis=1)
df_encoded = pd.concat([numerical_features, encoded_df, df['Exam_Score']], axis=1)


# Split the data into features and target variable
X = df_encoded.drop('Exam_Score', axis=1)
y = df_encoded['Exam_Score']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Fitting a linear regression model to the training data
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict the target variable for the testing data
y_pred = lr.predict(X_test)
```

```python
#model importing
from sklearn.linear_model import LinearRegression

# to split the data set into train and test sets and to do cross validation
from sklearn.model_selection import train_test_split, cross_val_score


#These are for evaluating the model accuracy
from sklearn.metrics import mean_squared_error, r2_score


# Evaluate the model performance using mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('R^2 Score:', r2)

# Perform k-fold cross-validation to evalute model
scores = cross_val_score(lr, X, y, cv=5, scoring='neg_mean_squared_error')
mse_cv = -1 * scores.mean()
print('Cross-Validated Mean Squared Error:', mse_cv)
```

### b. Compare different models based on the performance measures.

```python
## Regression Analysis: Mean Squared Error and R^2 Metric

# OLS
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

reg1 = LinearRegression().fit(X_train, y_train)
y_pred1 = reg1.predict(X_test)
mse_ols = mean_squared_error(y_test, y_pred1)
r2_ols = r2_score(y_test, y_pred1)
print("The MSE using OLS is:", mse_ols)
print("The R² using OLS is:", r2_ols)

# Ridge
from sklearn.linear_model import RidgeCV

reg2 = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1, 10, 100]).fit(X_train, y_train)
y_pred2 = reg2.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred2)
r2_ridge = r2_score(y_test, y_pred2)
print("The MSE using Ridge is:", mse_ridge)
print("The R² using Ridge is:", r2_ridge)

# Lasso
from sklearn.linear_model import LassoCV

reg3 = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1, 10, 100], random_state=0).fit(X_train, y_train)
y_pred3 = reg3.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred3)
r2_lasso = r2_score(y_test, y_pred3)
print("The MSE using Lasso is:", mse_lasso)
print("The R² using Lasso is:", r2_lasso)

```

### Model Comparison

   - R² Score: 0.7738

2. Ridge Regression:  
   - MSE: 3.1054

   - L2 regularization stabilizes the coefficients, reducing overfitting risk.  
   - Outperformed OLS slightly, indicating improved generalizability.

3. Lasso Regression:  
   - MSE: 3.1055  
   - R² Score: 0.77398  
   - L1 regularization shrinks some coefficients to zero, performing feature selection.  
   - Slightly behind Ridge in performance but comparable.  


### Final Selected Model: Ridge Regression
Reason: Ridge Regression had the best performance metrics with the lowest MSE and the highest R² score.
Regularization: Ridge’s L2 penalty ensures stable and robust coefficients, making the model more suitable for generalization to unseen data.
Ridge is preferred when small improvements in prediction accuracy are required without dropping variables to zero (as in Lasso).
